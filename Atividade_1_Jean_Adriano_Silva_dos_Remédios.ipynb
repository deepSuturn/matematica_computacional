1. A aplicação e a demonstração visual da rotação horária de 90°, seguida de shear (cisalhamento) em x, conforme o procedimento mostrado no slide 60 ao slide 68 da apresentação “Matrix multiplication as composition”.

R:

```python
import numpy as np
import matplotlib.pyplot as plt

rotacao = np.array([
    [0, 1],
    [-1, 0]
])

shear = np.array([
    [1, 3],
    [0, 1]
])

# VETORES DA BASE
vetor_b = np.array([[1,0], [0,1]])

# APLICAÇÃO DAS TRANSFORMAÇÕES
rot_dem = rotacao @ vetor_b
shear_dem = shear @ rot_dem

print("After rotation:")
print(rot_dem)
print("After shear:")
print(shear_dem)

# DEMONSTRAÇÃO VISUAL
fig, graf = plt.subplots(1, 3, figsize=(15, 5))

# Original
graf[0].quiver(0, 0, vetor_b[0,0], vetor_b[0,1], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label='î [1,0]')
graf[0].quiver(0, 0, vetor_b[1,0], vetor_b[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label='ĵ [0,1]')
graf[0].set_xlim(-3, 3)
graf[0].set_ylim(-3, 3)
graf[0].grid(True, alpha=0.3)
graf[0].set_title('Original')
graf[0].legend()
graf[0].set_aspect('equal')

#Após Rotação
graf[1].quiver(0, 0, rot_dem[0,0], rot_dem[1,0], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label=f'rot(î) [{rot_dem[0,0]},{rot_dem[0,1]}]')
graf[1].quiver(0, 0, rot_dem[0,1], rot_dem[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label=f'rot(ĵ) [{rot_dem[1,0]},{rot_dem[1,1]}]')
graf[1].set_xlim(-3, 3)
graf[1].set_ylim(-3, 3)
graf[1].grid(True, alpha=0.3)
graf[1].set_title('Após Rotação 90°')
graf[1].legend()
graf[1].set_aspect('equal')

#Após Shear
graf[2].quiver(0, 0, shear_dem[0,0], shear_dem[1,0], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label=f'shear(î) [{shear_dem[0,0]},{shear_dem[1,0]}]')
graf[2].quiver(0, 0, shear_dem[0,1], shear_dem[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label=f'shear(ĵ) [{shear_dem[0,1]},{shear_dem[1,1]}]')
graf[2].set_xlim(-3, 3)
graf[2].set_ylim(-3, 3)
graf[2].grid(True, alpha=0.3)
graf[2].set_title('Após Shear')
graf[2].legend()
graf[2].set_aspect('equal')

plt.tight_layout()
plt.show()
```

2. Mostre visualmente e explique o valor do determinante depois das transformações

```python
# CÁLCULO DOS DETERMINANTES
det_original = np.linalg.det(vetor_b)
det_rotacao = np.linalg.det(rot_dem)
det_shear = np.linalg.det(shear_dem)

# DEMONSTRAÇÃO VISUAL
fig, graf = plt.subplots(1, 3, figsize=(18, 5))

# Original
original_polygon = Polygon([[0, 0], vetor_b[0], vetor_b[0] + vetor_b[1], vetor_b[1]],
                          alpha=0.3, color='green', label=f'Area (det={det_original:.1f})')
graf[0].add_patch(original_polygon)
graf[0].quiver(0, 0, vetor_b[0,0], vetor_b[0,1], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label='î [1,0]')
graf[0].quiver(0, 0, vetor_b[1,0], vetor_b[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label='ĵ [0,1]')
graf[0].set_xlim(-3, 3)
graf[0].set_ylim(-3, 3)
graf[0].grid(True, alpha=0.3)
graf[0].set_title(f'Original\nDeterminante: {det_original:.1f}')
graf[0].legend()
graf[0].set_aspect('equal')

# Após Rotação
rot_polygon = Polygon([[0, 0], rot_dem[:,0], rot_dem[:,0] + rot_dem[:,1], rot_dem[:,1]],
                     alpha=0.3, color='orange', label=f'Area (det={det_rotacao:.1f})')
graf[1].add_patch(rot_polygon)
graf[1].quiver(0, 0, rot_dem[0,0], rot_dem[1,0], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label=f'rot(î) [{rot_dem[0,0]},{rot_dem[1,0]}]')
graf[1].quiver(0, 0, rot_dem[0,1], rot_dem[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label=f'rot(ĵ) [{rot_dem[0,1]},{rot_dem[1,1]}]')
graf[1].set_xlim(-3, 3)
graf[1].set_ylim(-3, 3)
graf[1].grid(True, alpha=0.3)
graf[1].set_title(f'Após Rotação 90°\nDeterminante: {det_rotacao:.1f}')
graf[1].legend()
graf[1].set_aspect('equal')

# Após Shear
shear_polygon = Polygon([[0, 0], shear_dem[:,0], shear_dem[:,0] + shear_dem[:,1], shear_dem[:,1]],
                       alpha=0.3, color='purple', label=f'Area (det={det_shear:.1f})')
graf[2].add_patch(shear_polygon)
graf[2].quiver(0, 0, shear_dem[0,0], shear_dem[1,0], angles='xy', scale_units='xy', scale=1,
               color='r', width=0.02, label=f'shear(î) [{shear_dem[0,0]},{shear_dem[1,0]}]')
graf[2].quiver(0, 0, shear_dem[0,1], shear_dem[1,1], angles='xy', scale_units='xy', scale=1,
               color='b', width=0.02, label=f'shear(ĵ) [{shear_dem[0,1]},{shear_dem[1,1]}]')
graf[2].set_xlim(-3, 3)
graf[2].set_ylim(-3, 3)
graf[2].grid(True, alpha=0.3)
graf[2].set_title(f'Após Shear\nDeterminante: {det_shear:.1f}')
graf[2].legend()
graf[2].set_aspect('equal')

plt.tight_layout()
plt.show()
```

Explicação do determinante: Como inicialmente a área do determinante é 1, após a rotação, vemos que o valor 1 informa que a orientação foi mantida e a área se mantém a mesma, após a segunda transformação o valor 1 informa que apesar do shear ter esticado o plano, a sua área se mantém a mesma..


3. Para o seu caso, faça o mesmo procedimento mostrado no slide 70 até o slide 92 da apresentação “Matrix multiplication as composition".

```python
#MULTIPLICAÇÃO DE MATRIZES DA FORMA COMO DEFINIDA NO MATERIAL
#[ae+bg af+bh]
#[ce+dg cf+dh]
def multiplicacao_matrizes(A: np.array, B: np.array) -> np.array:
    return np.array([
        [A[0,0]*B[0,0] + A[0,1]*B[1,0], A[0,0]*B[0,1] + A[0,1]*B[1,1]],
        [A[1,0]*B[0,0] + A[1,1]*B[1,0], A[1,0]*B[0,1] + A[1,1]*B[1,1]],
    ])


composicao = multiplicacao_matrizes(shear, rotacao)
print(composicao)
```

```python
composicao_inv = multiplicacao_matrizes(rotacao, shear)
print(composicao_inv)
```

5. Faça uma transformação linear em um vetor x de duas dimensões para obter um vetor y orientado a 90° em relação a x. Calcule o produto escalar entre os dois vetores. Interprete o resultado.

```python
#TRANSFORMAÇÃO DE ROTAÇÃO
rotacao_90_anti_horario = np.array([
    [0, -1],
    [1, 0]
])

x = np.array([3, 0])
print(x)

y = rotacao_90_anti_horario @ x
print(y)

dot = np.dot(x, y)
print(dot)
```

Interpretação: Qualquer produto escalar entre vetores que estão ortogonais entre si vai dar resultado 0, isso inclui quaisquer dois vetores que estão orientados a 90º um em relação ao outro, isso informa que eles não tem qualquer correlação linear entre si.
